```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold, cross_val_score, cross_val_predict
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# Load data from the CSV file
data = pd.read_csv("ies_scores.csv")
data1 = pd.read_csv("personality_data_classified.csv")

# Extract input features and output labels
X = data.iloc[:, 1:]  # Assuming the input features start from the 7th column (index 6)
y = data1.iloc[:, 1:6]  # Assuming the output aspects are from the 2nd to 6th column

# One-hot encode the categorical input features
# X_encoded = pd.get_dummies(X)

# Create a label encoder for the output labels
label_encoder = LabelEncoder()

# Perform K-fold cross-validation for each aspect separately
num_folds = 5
skf = StratifiedKFold(n_splits=num_folds, shuffle=True, random_state=42)

for aspect in y.columns:
    print(f"Aspect: {aspect}")
    aspect_labels = label_encoder.fit_transform(y[aspect])

    # Create a logistic regression model for the current aspect
    model = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=1000)

    # Perform K-fold cross-validation
    scores = cross_val_score(model, X, aspect_labels, cv=skf)

    # Print average accuracy and classification report for the current aspect
    print(f"Average Accuracy: {np.mean(scores):.2f}")

    model.fit(X, aspect_labels)  # Fit the model with the entire data before making predictions

    predictions = model.predict(X)
    print("Classification Report:")
    print(classification_report(aspect_labels, predictions))
    print("------------------------------")
```
